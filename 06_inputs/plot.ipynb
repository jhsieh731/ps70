```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
x = np.array([0, 10, 20, 29, 39, 48])
y = np.array([53812, 55465, 57074, 58477, 60419, 62148])

# Calculate the line of best fit
coefficients = np.polyfit(x, y, 1)
line_of_best_fit = np.poly1d(coefficients)

# Generate x values for the line of best fit
x_fit = np.linspace(x.min(), x.max(), 100)
y_fit = line_of_best_fit(x_fit)

# equation for line of best fit (to print)
equation = f'y = {coefficients[0]:.2f}x + {coefficients[1]:.2f}'

# Plot the points and the line of best fit
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x_fit, y_fit, color='red', label=f'Line of Best Fit ({equation})')

plt.xlabel('Mass (g)')
plt.ylabel('Summed proxy capacitance value')
plt.title('Sum of proxy capacitance vs. weight')
plt.legend()
plt.show()
```
